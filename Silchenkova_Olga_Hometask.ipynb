### Задание 4

Найдем аппроксимирующий полином. Для этого, используя МНК (метод наименьших квадратов), найдем коэффециенты для нашего аппроксимирующего полинома, по формуле:

$$ A = (M^T * M)^{-1}*M^T*Y,$$

где A - матрица полученных коэф., M  - матрица значений $x$ нашей "модельной" задачи, Y - матрица знаечний $y$ (Если мы представляем полнимо в виде $y = ax^0 + bx^1 + ...$ .

Для реализации МНК используем встроенную функцию $np.linalg.lstsq()$

```python
#k - степень будущего полинома
def model(x, y, k): 
    M = np.zeros((len(x), k+1))
    A = np.zeros(k)
    for i in range (k+1):
        for j in range (len(x)):
            M[j][i] = x[j] ** (k - i)
    A = np.linalg.lstsq(M, y, rcond=None)[0] #встроенный МНК
    return A
```

### Задание № 5

Для опредение СКО (средней квадратической ошибки) или , т.е. определение точность аппрокимируещего полинома, нам необходимо создать функцию для нахождения значений полином :

```python
def Polinom_value(a, x): # (a - матрица коэффцициентов) 
    p = 0.0
    for i in range(len(a)):
        p *= x
        p += a[i]
    return p
```

И теперь сама функция нахождения СКО :

$$ F(a,b) = \sum_{i = 1}^{n} (y_i - (ax_1^0 + bx_i^1 + ..))^2$$

В начале мы создаем значение ошибки $error$ с которой мы сначало сравниваем СКО - мы устанавливаем степень полинома $k = 1$. Если значение полученной суммы (при увеличении $k$) больше ошибки , то значит мы нашли оптимальную степень полинома (k = k - 1), иначе перезаписываем в $error$ значение полученной сум. и делаем опять все вычисления опять.

```python
# вычисление средне квадратическую ошибку, для нахождение оптимального k
def CKO(x, y):
    error = np.sum((y - Polinom_value(model(x,y,1), x))**2)
    summ = 0
    k = 0
    for i in range(2,100):
        k = i
        a = model(x,y,i)
        p = Polinom_value(a, x)
        for j in range(len(y)):
            summ += (y[j] - p[j])**2
        if summ > error:
            k = k - 1
            break
        else:
            error = summ
            summ = 0
    return k
```

В связи с тем ,что дата у нас представлена в форме $месяц - год$ , то каждому такому значению по очереди , начиная с единицы , обозначим номер $1 , 2 , 3, .., 13$ соответсвенно. 
Так же найдем значени оптимального $k$ для каждой сети:

```python
Date = np.arange(len(D))+1
cko_FB = CKO(Date, FB)
cko_Pr = CKO(Date, Pr)
cko_Tw = CKO(Date, Tw)
cko_Ins = CKO(Date, Ins)
cko_YT = CKO(Date, YT)
cko_red = CKO(Date, red)
cko_Tum = CKO(Date, Tum)
cko_VK = CKO(Date, VK)
cko_Lk = CKO(Date, Lk)
cko_Gg = CKO(Date, Gg)
cko_news = CKO(Date, news)
cko_Fark = CKO(Date, Fark)
cko_other = CKO(Date, other)
```

Дальше найдем МНК и значения самого полинома для каждой соц. сети :

```python
a1 = model(Date, FB, cko_FB)
p1 = Polinom_value(a1, Date)

a2 = model(Date, Pr, cko_Pr)
p2 = Polinom_value(a2, Date)

a3 = model(Date, Tw, cko_Tw)
p3 = Polinom_value(a3, Date)

a4 = model(Date, Ins, cko_Ins)
p4 = Polinom_value(a4, Date)

a5 = model(Date, YT, cko_YT)
p5 = Polinom_value(a5, Date)

a6 = model(Date, red, cko_red)
p6 = Polinom_value(a6, Date)

a7 = model(Date, Tum, cko_Tum)
p7 = Polinom_value(a7, Date)

a8 = model(Date, VK, cko_VK)
p8 = Polinom_value(a8, Date)

a9 = model(Date, Lk, cko_Lk)
p9 = Polinom_value(a9, Date)

a10 = model(Date, Gg, cko_Gg)
p10 = Polinom_value(a10, Date)

a11 = model(Date, news, cko_news)
p11 = Polinom_value(a11, Date)

a12 = model(Date, Fark, cko_Fark)
p12 = Polinom_value(a12, Date)

a13 = model(Date, other, cko_other)
p13 = Polinom_value(a13, Date)
```
